```python
# however, the above visualization poorly located each plot, so we are gonna do it differently, by using subplots

temp1=0
temp2=0
fig, ax = plt.subplots(5,2, sharey=True)

for i in range(1, len(df.store.unique())+1):
    ax[temp1,temp2].plot(df[df.store == i].date, df[df.store == i].sales)
    ax[temp1,temp2].set_title('Store # {}'.format(i))
    if temp2 == 1:
        temp2-=1
        temp1+=1
    elif temp2==0:
        temp2+=1
fig.tight_layout()
```
